```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
from SalesDataCleaner import SalesDataCleaner
```

```python
url = 'https://raw.githubusercontent.com/FrancescoMariottini/project3/main/inputs/all_sales_data.csv'
# url = './all_sales_data.csv'
sdc = SalesDataCleaner(url)
sdc.clean()
```

```python
sdc_df = sdc.get_cleaned_data()
sdc_df['price_by_area'] = sdc_df['price'] / sdc_df['area']
```

```python
# What are the less expensive municipalities in Belgium? (1-Average, 2-median, 3-price per square meter)
graph_data = sdc_df.groupby('postcode')['price'].mean().sort_values()[:5].to_frame().reset_index()
ax = sns.barplot(x='postcode', y='price', data=graph_data)
ax.set_title('The least expensive in Belgium?')
ax.set_ylabel('Average of price')
```

```python
graph_data = sdc_df.groupby('postcode')['price'].median().sort_values()[:5].to_frame().reset_index()
ax = sns.barplot(x='postcode', y='price', data=graph_data)
ax.set_title('The least expensive in Belgium?')
ax.set_ylabel('Median of price')
```

```python
graph_data = sdc_df.groupby('postcode')['price_by_area'].mean().sort_values()[:5].to_frame().reset_index()
ax = sns.barplot(x='postcode', y='price_by_area', data=graph_data)
ax.set_title('The least expensive in Belgium?')
ax.set_ylabel('Price per m2')
```

```python
# What are the less expensive municipalities in Wallonia? (1-Average, 2-median, 3-price per square meter)
graph_data = sdc_df[sdc_df.region=='W'].groupby('postcode')['price'].mean().sort_values()[:5].to_frame().reset_index()
ax = sns.barplot(x='postcode', y='price', data=graph_data)
ax.set_title('The least expensive in Wallonia?')
ax.set_ylabel('Average of price')
```
